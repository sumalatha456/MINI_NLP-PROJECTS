```python
train['Gender'] = train['Gender'].fillna(train['Gender'].mode()[0])
```

```python
train['Married'] = train['Married'].fillna(train['Married'].mode()[0])
```

```python
train['Dependents'] = train['Dependents'].fillna(train['Dependents'].mode()[0])
```

```python
train['Self_Employed'] = train['Self_Employed'].fillna(train['Self_Employed'].mode()[0])
```

```python
train['Credit_History'] = train['Credit_History'].fillna(train['Credit_History'].mode()[0])
```

```python
train['Loan_Amount_Term'] = train['Loan_Amount_Term'].fillna(train['Loan_Amount_Term'].mode()[0])
```

```python
train['LoanAmount'] = train['LoanAmount'].fillna(train['LoanAmount'].median())
```

```python
train.isna().sum()
```

```python
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
feature_Col = ['Gender', 'Married', 'Dependents', 'Education', 'Self_Employed', 'Property_Area', 'Loan_Status']
le = LabelEncoder()
for col in feature_Col:
  train[col] = le.fit_transform(train[col])
```

```python
train.Loan_Status = train.Loan_Status.replace({"Y": 1, "N": 0})
```

```python
train.head(3)
```

```python
train['total_income'] = train['ApplicantIncome'] + train['CoapplicantIncome']
```

```python
train.drop(columns=['ApplicantIncome', 'CoapplicantIncome'], inplace=True)
```

```python
X = train.drop(columns=['Loan_Status'])
y = train['Loan_Status']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=7)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=100,max_features ='sqrt')
rf.fit(X_train, y_train)
rf.pred=rf.predict(X_test).astype(int)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(classification_report(y_test, rf.pred))
confusion_matrix(y_test, rf.pred)
accuracy_score(y_test, rf.pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
dt.pred=dt.predict(X_test).astype(int)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
lr.pred=lr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
print(classification_report(y_test, lr.pred))
confusion_matrix(y_test, lr.pred)
accuracy_score(y_test, lr.pred)
```
