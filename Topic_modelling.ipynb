## Topic modeling is an unsupervised technique
Two methods: latent Dirichlet allocation and non negative matrix factorization.

```python
import pandas as pd
import numpy as np

# Adding the 'error_bad_lines=False' parameter to skip problematic lines.
# and quoting=3 to handle unescaped quotes.
reviews_datasets = pd.read_csv('Reviews.csv', on_bad_lines='skip', quoting=3)

reviews_datasets = reviews_datasets.head(20000)
reviews_datasets.dropna(inplace=True)
```

```python
!pip install scikit-learn
from sklearn.feature_extraction.text import CountVectorizer # Corrected the typo from 'CounytfVectorizer' to 'CountVectorizer'
count_vect = CountVectorizer(max_df=0.8, min_df=2, stop_words='english')
doc_term_matrix = count_vect.fit_transform(reviews_datasets['Text'].values.astype('U'))
```

```python
CountVectorizer?
```

```python
print(doc_term_matrix)
```

```python
from sklearn.decomposition import LatentDirichletAllocation
LDA = LatentDirichletAllocation(n_components=5, random_state=42)
LDA.fit(doc_term_matrix)
```

```python
pip list
```

```python
import random
for i in range(10):
    random_id = random.randint(0,len(count_vect.get_feature_names_out()))
    print(count_vect.get_feature_names_out()[random_id])
```

```python
first_topic = LDA.components_[2]
```

```python
top_topic_words = first_topic.argsort()[-10:] ## last 10 words
```

```python
for i in top_topic_words:
    print(count_vect.get_feature_names_out()[i])
```

```python
for i,topic in enumerate(LDA.components_):
    print(f'Top 10 words for topic #{i}:')
    print([count_vect.get_feature_names_out()[i] for i in topic.argsort()[-10:]])
    print('\n')
```
